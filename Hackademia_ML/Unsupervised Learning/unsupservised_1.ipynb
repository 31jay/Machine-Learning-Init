```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
import random 
import seaborn as sns

students = [chr(i+64) for i in range(1,100)]
math_marks = [random.randrange(1,100) for i in range(len(students))]
science_marks = [random.randrange(1,100) for i in range(len(students))]

print(len(students),len(math_marks),len(science_marks))
data_frame = pd.DataFrame({
    'Names':students,
    'Maths':math_marks,
    'Science':science_marks
})

display(data_frame)
```

```python
plt.scatter(data_frame.Maths,data_frame.Science)
plt.xlabel('Maths')
plt.ylabel('Science')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)

kmeans.fit(data_frame[['Maths','Science']])
data_frame['cluster'] = kmeans.labels_
display(data_frame)
```

```python
colors = {0: 'RED', 1:'BLUE', 2:'GREEN', 3:'YELLOW', 4:'PURPLE',5:'ORANGE'}
plt.figure(figsize=(10, 10))
for cluster in data_frame['cluster'].unique():
    plt.scatter(data_frame[data_frame['cluster'] == cluster]['Maths'], data_frame[data_frame['cluster'] == cluster]['Science'], c=colors[cluster])
plt.xlabel('Maths')
plt.ylabel('Science')
plt.title('Scatter Plot of Maths vs Science with K-Means Clustering')
plt.legend() 
plt.show()
```
