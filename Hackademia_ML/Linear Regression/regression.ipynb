```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_student = pd.read_csv('student_scores.csv')
display(data_student.head(6))
```

2. Visualizing Data

```python
x=data_student.Hours
y=data_student.Scores
plt.scatter(x,y)
```

3. Using Least Squre Regression

```python
def compute_n_c(x,y):
    x_mean=sum(x)/len(x)
    y_mean=sum(y)/len(y)
    numerator,denominator= 0,0 
    for i in range(len(x)):
        numerator+=(x[i]-x_mean)*(y[i]-y_mean)
        denominator+= (x[i]-x_mean)*(x[i]-x_mean)
    
    m=numerator/denominator
    c=y_mean-m*x_mean
    return m,c

slope,intercept=compute_n_c(x,y)
print(slope,intercept)
```

**4. Plotting best fit line in the graph**

Linear Regression Line

```python
y_predicted = [slope*i+intercept for i in x]
plt.scatter(x,y)
plt.plot(x,np.array(y_predicted),color='red')
```

**5. R Squared Method**

R squared is computed to get how well the regression line justifies the data.
R squared value is a statistical measure of how close the data points are to the fitted regression line. R square ranges from 0 to 1. The closer the value of R square is to 1, the better the model is.

R=1-(SSR/SST)
SSR = sum of square of residuals = sum of square of (y_predicted - y_actual)  
SST = sum of square of total mean = sum of square of (y_actual - y_mean)  

R^2 = 1 : All the variation is explained by the model  
R^2 = 0 : No variation is explained by the model  
R^2 = 0.83 : 83% variation is explained by the model  





```python
def compute_R_square(x,y,y_predicted):
    numerator,denominator=0,0
    y_mean=sum(y)/len(y)
    for i in range(len(x)):
        numerator+=(y[i]-y_predicted[i])**2
        denominator+=(y[i]-y_mean)**2
    r_square=numerator/denominator
    return r_square
r_square=compute_R_square(x,y,y_predicted)
print(r_square)
```

**6. Checking Model in Unseen data**  
Trying for a completely unseen data (4 hours, 8 hours, 15 hours)

```python
hours_test=[4,8,15]
predicted_scores=[slope*i+intercept for i in hours_test]

print(predicted_scores)
```
